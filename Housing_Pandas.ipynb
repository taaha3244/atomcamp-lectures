#Grouping Data

```python
#Group by 'ocean_proximity' and get median of all numerical columns
group_sum = housing.groupby('ocean_proximity').mean()
group_sum
```

```python
#Count the number of entries in each 'ocean_proximity' category
group_count = housing.groupby('ocean_proximity').size()


group_count2=housing['ocean_proximity'].value_counts()

group_count2
```

```python
#  Calculate average 'median_house_value' by 'ocean_proximity'
group_mean = housing.groupby('ocean_proximity')['median_house_value'].mean()
group_mean
```

##Grouping with Multiple Columns

```python
#Group by both 'ocean_proximity' and 'housing_median_age', and calculate mean values
multi_group_mean = housing.groupby(['ocean_proximity', 'housing_median_age'])['housing_median_age'].mean()
multi_group_mean
```

```python
# Get descriptive statistics for 'median_house_value' grouped by 'ocean_proximity'
group_description = housing.groupby('ocean_proximity')['housing_median_age'].describe()
group_description
```

```python
# Calculate the median of 'total_rooms' and mean of 'total_bedrooms' grouped by 'ocean_proximity'
custom_agg = housing.groupby('ocean_proximity').agg({
    'total_rooms': 'median',
    'total_bedrooms': 'mean'
})
custom_agg
```

#Joining Data frames

```python
near_bay_homes = housing[housing['ocean_proximity'] == 'NEAR BAY']
other_homes = housing[housing['ocean_proximity'] != 'NEAR BAY']
```

```python
# Concatenating the DataFrames vertically
concatenated_df = pd.concat([near_bay_homes, other_homes], axis=0)
concatenated_df
```

```python
# Mapping old values to new ones
proximity_map = {
    'NEAR BAY': 'NBAY',
    'INLAND': 'INL',
    'NEAR OCEAN': 'NOCEAN',
    '<1H OCEAN': 'OCEAN1H',
    'ISLAND': 'ISL'
}

# Use map to apply the changes
housing['ocean_proximity_short'] = housing['ocean_proximity'].map(proximity_map)

# Show the result
print(housing[['ocean_proximity', 'ocean_proximity_short']].head())
```
