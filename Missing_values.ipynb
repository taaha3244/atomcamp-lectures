#LOC & iLOC

Pandas loc and iloc
loc and iloc are methods in pandas for selecting rows and columns from a DataFrame.

loc
loc is label-based. You can access a group of rows and columns by labels or a boolean array.
Syntax: df.loc[row_labels, column_labels]

```python
df.loc[0]
```

```python
# Select rows with index labels 0 and 1, and columns 'Name' and 'Age'
df.loc[[0, 1], ['Name', 'Age']]
```

```python
# Select rows where 'Age' is greater than 30
df.loc[df['Age'] > 30]
```

```python
female_high_fare = df.loc[(df['Sex'] == 'female') & (df['Fare'] > 50)]
```

iloc
iloc is integer position-based. You can access rows and columns by their integer position.
Syntax: df.iloc[row_indices, column_indices]

```python
# Select the first row (index position 0)
df.iloc[0]
```

```python
# Select the first two rows and the first three columns
df.iloc[0:2, 0:3]
```

```python
# Select the last row
df.iloc[-1]
```

cut and qcut
cut
cut is used to segment and sort data values into bins. It is useful for converting continuous data into categorical data.

```python
subset = df.iloc[[1, 3, 5], [2, 4, 6]]
```

```python
df['Age_bin'] = pd.cut(df['Age'], bins=3)
```

```python
bins = [0, 18, 35, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df
```


qcut
qcut is used to segment and sort data values into bins based on sample quantiles. It ensures each bin has approximately the same number of observations.

```python
df['Fare_quantile'] = pd.qcut(df['Fare'], q=4)
```

```python
quantiles = [0, 0.25, 0.5, 0.75, 1.0]
labels = ['Low', 'Medium', 'High', 'Very High']
df['Fare_group'] = pd.qcut(df['Fare'], q=quantiles, labels=labels)
```

```python
y1 = updated_df['Survived']
updated_df.drop("Survived",axis=1,inplace=True)
from sklearn import metrics
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(updated_df,y1,test_size=0.3)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
print(metrics.accuracy_score(pred,y_test))
```
